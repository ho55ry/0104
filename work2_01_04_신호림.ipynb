```python
# 모듈 불러오기
import pandas as pd
```

```python
# 파일 => DF
FILE='../datafiles/남북한발전전력량.xlsx'
df=pd.read_excel(FILE)
```

```python
# 데이터 확인
print(df.shape)
df.info()
```

```python
df.columns
```

```python
# 이름 쉽게 바꿈
df.columns=['location', 'sort', '1990', '1991', '1992', '1993', '1994', '1995',
       '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004',
       '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
       '2014', '2015', '2016']
```

```python
df.shape
```

```python
# 북한 신재생 row 추가해주기
A=['북한','신재생']; [A.append('-') for _ in range(27)]
df.loc[9]=A
```

```python
# 결측치 임의 처리
df.iloc[[1,2,3,4],0]='남한'
df.iloc[[6,7,8],0]='북한'
```

```python
# 결측치 '-' => 0으로 바꿔주기
df.replace('-',0,inplace=True)
```

```python
# 인덱스 바꿔주기
df.set_index(['location','sort'])
```

```python
# 행 평균 값 확인
df.mean(axis=1)
```

```python
# 행 평균값 컬럼 만들기
df['평균값']=df.mean(axis=1).round()
```
